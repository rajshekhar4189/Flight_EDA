```python
final_df=[Test_set_df,Data_train_df]
final_df=pd.concat(final_df)
```

```python
##Feature Engineering
```

```python
final_df['Date']=final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month']=final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year']=final_df['Date_of_Journey'].str.split('/').str[2]
final_df
```

```python
final_df['Date']=pd.to_numeric(final_df['Date'])
final_df['Month']=pd.to_numeric(final_df['Month'])
final_df['Year']=pd.to_numeric(final_df['Year'])
final_df.info()
```

```python
final_df.drop('Date_of_Journey',axis=1,inplace=True)
final_df
```

```python
final_df.head(1)
```

```python
final_df['Arrival_Time']=final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df.head()
```

```python
final_df.isnull().sum()
```

```python
final_df['Arrival_Hour']=final_df['Arrival_Time'].str.split(':').str[0]
final_df['Arrival_Min']=final_df['Arrival_Time'].str.split(':').str[1]
final_df
```

```python
final_df['Arrival_Hour']=pd.to_numeric(final_df['Arrival_Hour'])
final_df['Arrival_Min']=pd.to_numeric(final_df['Arrival_Min'])
```

```python
final_df.info()
```

```python
final_df.head()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df['Total_Stops']=final_df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```
